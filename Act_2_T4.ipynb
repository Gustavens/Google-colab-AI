En este paso clave del aprendizaje automático, trabajamos con un conjunto de datos sobre vinos y lo dividimos en dos partes con propósitos distintos:
Conjunto de Entrenamiento: Es el grupo de datos que utilizamos para enseñar al modelo. Básicamente, el algoritmo estudia los patrones y relaciones entre los 13 atributos de cada vino y su clasificación para poder predecir correctamente a futuro.

Conjunto de Prueba: Estos datos se dejan aparte y no se usan en el entrenamiento. Su función es evaluar el rendimiento del modelo una vez que ha aprendido, ayudándonos a medir qué tan bien puede generalizar sus predicciones a datos nuevos y desconocidos, simulando así una aplicación en el mundo real.

Esta separación es fundamental para asegurarnos de que el modelo no solo "memoriza" los datos de entrenamiento, sino que realmente aprende a reconocer patrones para hacer predicciones acertadas.


**Genera el modelo utilizando el conjunto de datos de entrenamiento.**

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # Importamos el clasificador de Árbol de Decisión
```

```python
# Cargar y preparar los datos (reutilizando el código anterior para asegurar la consistencia)
wine = load_wine()
X = wine.data
y = wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=109)

print("--- Paso 4: Generando el modelo de Árbol de Decisión con el conjunto de entrenamiento ---")

# 1. Generar el modelo utilizando el conjunto de datos de entrenamiento
# Creamos una instancia del clasificador de Árbol de Decisión
model = DecisionTreeClassifier(random_state=109) # Se usa random_state para reproducibilidad del árbol

# Entrenamos el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

print("Modelo de Árbol de Decisión entrenado exitosamente.")

#El algoritmo de DecisionTreeClassifier ahora "aprende" las reglas y patrones de decisión a partir de estos datos para poder clasificar nuevos vinos.
```

```python
print("\n--- Paso 5: Realizando predicciones con el conjunto de datos de prueba ---")

# 2. Con el modelo construido, realiza predicciones utilizando el conjunto de datos de prueba
# y almacena los resultados en una variable nombrada y_pred.
y_pred = model.predict(X_test)

print("Predicciones realizadas con éxito.")
print(f"Primeras 10 predicciones: {y_pred[:10]}")
print(f"Primeras 10 etiquetas reales: {y_test[:10]}")

print("\nEl modelo ha sido entrenado y las predicciones han sido generadas.")
```

y_pred = model.predict(X_test):

Esta es la fase de predicción.
Una vez que el modelo está entrenado, usamos el método model.predict().

Le pasamos X_test (los atributos del conjunto de prueba, datos que el modelo nunca ha visto durante el entrenamiento).

El modelo utiliza las reglas que aprendió para predecir el tipo de vino para cada muestra en X_test.

Los resultados de estas predicciones se almacenan en la variable y_pred.

Evaluación del modelo.

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics # Importamos el módulo metrics

# Cargar y preparar los datos (reutilizando el código anterior para asegurar la consistencia)
wine = load_wine()
X = wine.data
y = wine.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=109)

# Generar y entrenar el modelo de Árbol de Decisión
model = DecisionTreeClassifier(random_state=109)
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)

print("--- Evaluación inicial del modelo de Árbol de Decisión ---")
# Evaluar el modelo utilizando accuracy_score
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy del modelo de Árbol de Decisión: {accuracy:.4f}")

# También podemos ver otras métricas para una evaluación más completa
print("\nOtras métricas de evaluación:")
print(f"Precision: {metrics.precision_score(y_test, y_pred, average='weighted'):.4f}")
print(f"Recall: {metrics.recall_score(y_test, y_pred, average='weighted'):.4f}")
print(f"F1-Score: {metrics.f1_score(y_test, y_pred, average='weighted'):.4f}")
print("\nMatriz de Confusión:\n", metrics.confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:\n", metrics.classification_report(y_test, y_pred, target_names=wine.target_names))
```

**Mejoras a considerar**

Reducir la complejidad del Árbol de Decisión, ajustando parámetros como max_depth o min_samples_leaf, lo que ayuda a evitar el sobreajuste.

Usar GridSearchCV, una herramienta que busca automáticamente la mejor combinación de parámetros para optimizar el rendimiento del modelo.

Probar con modelos de ensamble, como RandomForestClassifier, que combinan varios árboles para lograr predicciones más robustas y precisas.

Al entrenar un modelo con DecisionTreeClassifier, hay varios parámetros que puedes ajustar para mejorar su rendimiento y evitar que se sobreentrene.

Uso de DecisionTreeClassifier

```python
# Imprimimos un encabezado para indicar que vamos a mejorar el modelo
print("\n--- Mejorando el modelo de Árbol de Decisión (Ajuste de Hiperparámetros) ---")

#  Modelo 1: Limitando la profundidad del árbol para evitar sobreajuste
# Un árbol muy profundo puede memorizar los datos de entrenamiento, lo que reduce su capacidad de generalizar
model_tuned_1 = DecisionTreeClassifier(max_depth=5, random_state=109)  # Limita la profundidad máxima a 5 niveles
model_tuned_1.fit(X_train, y_train)  # Entrenamos el modelo con los datos de entrenamiento
y_pred_tuned_1 = model_tuned_1.predict(X_test)  # Hacemos predicciones sobre los datos de prueba
accuracy_tuned_1 = metrics.accuracy_score(y_test, y_pred_tuned_1)  # Calculamos la precisión del modelo
print(f"Accuracy del Árbol de Decisión (max_depth=5): {accuracy_tuned_1:.4f}")

#  Modelo 2: Estableciendo un número mínimo de muestras por hoja
# Esto evita que el árbol cree nodos hoja con muy pocos datos, lo que también ayuda a reducir el sobreajuste
model_tuned_2 = DecisionTreeClassifier(min_samples_leaf=5, random_state=109)  # Requiere al menos 5 muestras por hoja
model_tuned_2.fit(X_train, y_train)
y_pred_tuned_2 = model_tuned_2.predict(X_test)
accuracy_tuned_2 = metrics.accuracy_score(y_test, y_pred_tuned_2)
print(f"Accuracy del Árbol de Decisión (min_samples_leaf=5): {accuracy_tuned_2:.4f}")

#  Modelo 3: Búsqueda automática de los mejores parámetros con GridSearchCV
# Esta técnica prueba múltiples combinaciones de parámetros para encontrar la que da mejor rendimiento
from sklearn.model_selection import GridSearchCV

# Definimos el espacio de búsqueda de parámetros
param_grid = {
    'max_depth': [3, 5, 7, 10],  # Profundidades posibles del árbol
    'min_samples_leaf': [1, 3, 5, 10],  # Tamaños mínimos de hojas
    'criterion': ['gini', 'entropy']  # Funciones para medir la calidad de las divisiones
}

# Configuramos GridSearchCV con validación cruzada de 5 pliegues
grid_search = GridSearchCV(DecisionTreeClassifier(random_state=109), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)  # Entrenamos el modelo con todas las combinaciones posibles
```

```python
# Mostramos los mejores parámetros encontrados
print(f"\nMejores parámetros encontrados por GridSearchCV: {grid_search.best_params_}")

# Evaluamos el mejor modelo encontrado
best_tree_model = grid_search.best_estimator_
y_pred_best_tree = best_tree_model.predict(X_test)
accuracy_best_tree = metrics.accuracy_score(y_test, y_pred_best_tree)
print(f"Accuracy del Árbol de Decisión optimizado con GridSearchCV: {accuracy_best_tree:.4f}")

# Mostramos un reporte detallado de clasificación
print("\nReporte de Clasificación del modelo optimizado:\n", metrics.classification_report(y_test, y_pred_best_tree, target_names=wine.target_names))
```

**Explorando Predicciones y Complementación de Texto con BERT y GPT**

```python
from transformers import pipeline
import os

# --- Modelos para español ---
# Para generar texto (similar a GPT): usaremos 'dccuchile/bert-base-spanish-wwm-cased' o 'dccuchile/bert-base-spanish-wwm-cased-finetuned-sqac'

# Inicializar pipelines
# Para generación de texto en español (GPT-like)

try:
    generator_es = pipeline('text-generation', model='dccuchile/gpt2-small-spanish')
except Exception as e:
    print(f"Error initializing Spanish text generation pipeline: {e}")
    print("Please ensure you are authenticated with Hugging Face Hub.")
    print("Run `huggingface-cli login` in your terminal or pass your token to the pipeline function.")
    # Decide how to handle the error if the pipeline cannot be initialized
    # For example, you might skip the Spanish generation part or use a different model.
    generator_es = None # Set to None to indicate failure


# Para rellenar huecos en español (BERT-like)
try:
    fill_mask_es = pipeline('fill-mask', model='dccuchile/bert-base-spanish-wwm-cased')
except Exception as e:
    print(f"Error initializing Spanish fill-mask pipeline: {e}")
    print("Please ensure you are authenticated with Hugging Face Hub.")
    fill_mask_es = None # Set to None

# --- Modelos para inglés ---
# Para generación de texto (GPT-like)
try:
    generator_en = pipeline('text-generation', model='gpt2') # GPT-2 base model
except Exception as e:
     print(f"Error initializing English text generation pipeline: {e}")
     generator_en = None # Set to None


# Para rellenar huecos (BERT-like)
try:
    fill_mask_en = pipeline('fill-mask', model='bert-base-uncased')
except Exception as e:
    print(f"Error initializing English fill-mask pipeline: {e}")
    fill_mask_en = None # Set to None


# --- Textos/Oraciones de prueba ---
textos_espanol = [
    "El perro ladró fuertemente y el gato se ",
    "La inteligencia artificial transformará la sociedad de maneras ",
    "El cambio climático es un desafío global que requiere ",
    "En la cima de la montaña, la vista era ",
    "El futuro de la tecnología es prometedor, pero también presenta "
]

textos_ingles = [
    "The dog barked loudly and the cat ",
    "Artificial intelligence will transform society in ways ",
    "Climate change is a global challenge that requires ",
    "At the top of the mountain, the view was ",
    "The future of technology is promising, but also presents "
]

print("--- Predicciones y Complementación (Español) ---")
if generator_es or fill_mask_es: # Only proceed if at least one Spanish model loaded
    for i, text in enumerate(textos_espanol):
        print(f"\n--- Texto {i+1} (Español) ---")
        print(f"Original: '{text}'")

        # Predicción con modelo generativo (GPT-like)
        if generator_es:
            print("\n[Generación GPT-like]:")
            try:
                # Ajustamos max_length para controlar la longitud de la generación
                # no_repeat_ngram_size para evitar repeticiones
                # num_return_sequences para obtener varias opciones
                generated_text_es = generator_es(text, max_length=len(text.split()) + 10, num_return_sequences=1, do_sample=True, temperature=0.7)
                for j, gen_seq in enumerate(generated_text_es):
                    print(f"  Opción {j+1}: '{gen_seq['generated_text']}'")
            except Exception as e:
                print(f"  Error during Spanish GPT-like generation: {e}")


        # Predicción con modelo de rellenar huecos (BERT-like)
        if fill_mask_es:
            text_masked = text + " [MASK]"
            print("\n[Rellenar Huecos BERT-like (palabra final)]: ")
            try:
                filled_mask_es = fill_mask_es(text_masked)
                for j, fill in enumerate(filled_mask_es[:3]): # Mostrar las top 3 predicciones
                    print(f"  Opción {j+1}: '{text_masked.replace('[MASK]', fill['token_str'])}' (score: {fill['score']:.4f})")
            except Exception as e:
                 print(f"  Error during Spanish BERT-like fill-mask: {e}")
else:
    print("\nSkipping Spanish predictions due to model loading errors.")


print("\n\n" + "="*50 + "\n") # Separador


print("--- Predicciones y Complementación (Inglés) ---")
if generator_en or fill_mask_en: # Only proceed if at least one English model loaded
    for i, text in enumerate(textos_ingles):
        print(f"\n--- Texto {i+1} (Inglés) ---")
        print(f"Original: '{text}'")

        # Predicción con modelo generativo (GPT-like)
        if generator_en:
            print("\n[Generación GPT-like]:")
            try:
                generated_text_en = generator_en(text, max_length=len(text.split()) + 10, num_return_sequences=1, do_sample=True, temperature=0.7)
                for j, gen_seq in enumerate(generated_text_en):
                    print(f"  Opción {j+1}: '{gen_seq['generated_text']}'")
            except Exception as e:
                print(f"  Error during English GPT-like generation: {e}")

        # Predicción con modelo de rellenar huecos (BERT-like)
        if fill_mask_en:
            text_masked = text + " [MASK]"
            print("\n[Rellenar Huecos BERT-like (palabra final)]: ")
            try:
                filled_mask_en = fill_mask_en(text_masked)
                for j, fill in enumerate(filled_mask_en[:3]): # Mostrar las top 3 predicciones
                    print(f"  Opción {j+1}: '{text_masked.replace('[MASK]', fill['token_str'])}' (score: {fill['score']:.4f})")
            except Exception as e:
                print(f"  Error during English BERT-like fill-mask: {e}")
else:
     print("\nSkipping English predictions due to model loading errors.")
```
