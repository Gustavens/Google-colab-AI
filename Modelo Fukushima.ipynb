```python
import numpy as np

#Matriz que define al patron + como entrada
img_entrada = np.array([[0, 0, 0],
                        [1, 1, 1],
                        [0, 1, 0]])

#Definicion de los vectores para los pesos
#en lienas horizontales y verticales
pesos_horizontales = np.array([[-1, -1, -1],
                               [1, 1, 1],
                               [-1, -1, -1]])

pesos_verticales = np.array([[-1, 1, -1],
                             [-1, 1, -1],
                             [-1, 1, -1]])

def fn_activacion(x):
  return np.heaviside(x,0.5)

def celda_s(img_entrada, pesos):
  entrada_neta = np.dot(pesos.flatten(), img_entrada.flatten())
  print(entrada_neta)
  salida = fn_activacion(entrada_neta)
  return salida.reshape(1)

def celda_c(s_salida_horizontal, s_salida_vertical):
  vector_ent = np.concatenate([s_salida_horizontal, s_salida_vertical])
  entarda_neta = np.dot(vector_ent, np.array([-0.5,1]))
  salida = fn_activacion(entarda_neta)
  return salida

#C치lculo de la salida de una celda S
s_salida_horizontal = celda_s(img_entrada[:3, :], pesos_horizontales)
s_salida_vertical = celda_s(img_entrada[: , :3], pesos_verticales)

#C치lculo de la salida C
salida_c = celda_c(s_salida_horizontal, s_salida_vertical)

#Salidas
print("Imagen de entrada:\n",img_entrada)
print("Salida de celda S horizontal:\n", s_salida_horizontal)
print("Salida de celda S vertical:\n", s_salida_vertical)
print("Salida de celda C:\n",salida_c)
```
