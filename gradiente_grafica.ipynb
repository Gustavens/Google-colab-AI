```python
import numpy as np
import matplotlib.pyplot as plt

#Funcion de perdida
def loss_function(w):
  return (w - 4) ** 2 + 5

#Gradiente de la funcion de perdida
def gradient(w):
  return 2 * (w - 4)

#Tasa de aprendizaje
lr = 0.9
#peso
w = 0

#Inicializacion
w_values = [w]
loss_values = [loss_function(w)]

#criterio de paro
max_iterations = 50
tolerance = 1e-6

#ejecucion del graiente
for i in range(max_iterations):
  #cálculo del gradiente de la función de pérdida en el valor actual
  grad= gradient(w)

  #actualizacion
  w = w - lr * grad

  #agregar los valores al arreglo
  w_values.append(w)
  loss_values.append(loss_function(w))

  #verificar con el nivel de tolerancia
  if abs(loss_values[-1] - loss_values[-2]) < tolerance:
    break
```

```python
#grafico del DG
w_range = np.linspace(-5,5, 100)
loss_range = loss_function(w_range)
plt.plot(w_range, loss_range, label="Funcion de perdida")
plt.plot(w_values, loss_values,"-o", label='Iteraciones del descenso de gradiente')
plt.xlabel('w')
plt.ylabel("Valores de la funcion de perdida")
plt.legend()
plt.show()
```
