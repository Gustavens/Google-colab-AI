```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf


X = np.linspace(0, 6*np.pi,1000).reshape(-1,1)
y = np.sin(X)
#y= 2*X +0.5*np.sin(5*X)+0.2*X**2

model = Sequential()
model.add(Dense(  , activation='', input_shape=(1,)))
model.add(Dense(1, activation=''))

model.summary()
model.compile( loss=tf.keras.losses.MeanSquaredError(),  optimizer=tf.keras.optimizers.Adam(learning_rate = 0.1), metrics=['accuracy'] )
h = model.fit(X,y, batch_size=128, epochs=350)
y_pred = model.predict(X)
plt.plot(X, y, label='Real')
plt.plot(X, y_pred, label='Prediccion')
plt.show()
```
